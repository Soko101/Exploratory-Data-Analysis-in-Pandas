```python
#Example use case: Are there certain continents that have grown faster than others and in which ways
# we can use population since we have almost a 50 year span for countries

numeric_cols = df.select_dtypes(include='number').columns
continent_means = df.groupby('Continent')[numeric_cols].mean().sort_values(by="2022 Population",ascending=False)
continent_means
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
# Step 1: List the population columns you want
pop_cols = ['1970 Population', '1980 Population', '1990 Population', '2000 Population',
            '2010 Population', '2015 Population', '2020 Population', '2022 Population']

# Step 2: Group and calculate mean on just those columns
df2 = df.groupby('Continent')[pop_cols].mean().sort_values(by='2022 Population', ascending=False)

df2
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot(figsize=(20,7))
```
